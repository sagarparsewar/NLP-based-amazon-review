# Reading the CSV file :

```python
df_reviews_ratings = pd.read_csv('Amazon_reviews.csv')
df_reviews_ratings.head(5)
```

```python
df_reviews_ratings.isnull().sum()
```

```python
df_reviews_ratings=df_reviews_ratings.dropna()
```

```python
df_reviews_ratings=df_reviews_ratings.reset_index()
```

```python
df_reviews_ratings.columns = df_reviews_ratings.columns.str.strip()
```

```python
# train test split 
num = int(0.9*df_reviews_ratings.shape[0])
train = df_reviews_ratings[:num]
print(train.shape)
test = df_reviews_ratings[num:]
print(test.shape)


# defining the other parameters : 
target = 'Ratings'
sample_submission = 'submission'
```

```python
#train
```

# Auto_ViML : 

```python
model, features, trainm, testm = Auto_ViML(
                                            train,
                                            target,
                                            test,
                                            sample_submission ,
                                            hyper_param="GS",
                                            feature_reduction=False,     # Since we have less than 10 features
                                            scoring_parameter="balanced_accuracy",   # For classification choose 'balanced_accuracy'
                                            KMeans_Featurizer=True,     # For NLP and CatBoost
                                            Boosting_Flag=True,          # None : linear model , False : RandomForest , True : XGBoost , CatBoost : complex models
                                            Binning_Flag=False,        # For CatBoost
                                            Add_Poly=False,
                                            Stacking_Flag=False,
                                            Imbalanced_Flag=True,      # True if the data is imbalanced
                                            verbose=1,
)
```
