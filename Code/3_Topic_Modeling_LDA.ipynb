```python
import pandas as pd
import numpy as np
```

# Reading CSv file :

```python
df_reviews_ratings = pd.read_csv('PreProcessed_data.csv')
df_reviews_ratings.head(5)
```

### Removing null values if any :

```python
df_reviews_ratings.isnull().sum()
```

```python
df_reviews_ratings=df_reviews_ratings.dropna()
```

```python
df_reviews_ratings.isnull().sum()
```

```python
df_reviews_ratings.reset_index()
```

# Create Document Term Matrix :

### Topic modleing for Rating : 1 

```python
df_reviews_1_ratings= df_reviews_ratings[df_reviews_ratings.Ratings == 1]
```

```python
df_reviews_1_ratings.shape
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
data_cv = cv.fit_transform(df_reviews_1_ratings.Reviews)
```

```python
data_cv.shape
```

```python
data_stop = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_stop.index = df_reviews_1_ratings.index
```

```python
data_stop
```

### Modeling using LDA :

```python
import gensim
from gensim import matutils, models
import scipy.sparse
```

```python
tdm = data_stop.transpose()
tdm.head()
```

```python
# We're going to put the term-document matrix into a new gensim format, from df --> sparse matrix --> gensim corpus
sparse_counts = scipy.sparse.csr_matrix(tdm)
corpus = matutils.Sparse2Corpus(sparse_counts)
```

```python
id2word = dict((v, k) for k, v in cv.vocabulary_.items())
```

```python
lda = models.LdaModel(corpus=corpus, id2word=id2word, num_topics=3, passes=10)
lda.print_topics()
```

```python
lda.get_topics()
```
